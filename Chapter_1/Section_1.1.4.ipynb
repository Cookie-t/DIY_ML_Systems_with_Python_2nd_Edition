```python
# 导入pandas工具包，并将其重命名为pd。
import pandas as pd

'''
调用pandas工具包的read_csv函数，从数据文件breast-cancer-train.csv中读取训练样本;
并且选取'Clump Thickness', 'Cell Size', 'Type'三列，存储到变量train。
'''
train = pd.read_csv('datasets/breast-cancer-train.csv', usecols=['Clump Thickness', 'Cell Size', 'Type'])

# 选取'Clump Thickness', 'Cell Size'两列作为特征，构建训练特征样本。
X_train = train[['Clump Thickness', 'Cell Size']]
# 选取'Type'列数据为监督学习的类别目标。
y_train = train[['Type']]


'''
调用pandas工具包的read_csv函数，从数据文件breast-cancer-test.csv中读取测试样本;
并且选取'Clump Thickness', 'Cell Size', 'Type'三列，存储到变量test。
'''
test = pd.read_csv('datasets/breast-cancer-test.csv', usecols=['Clump Thickness', 'Cell Size', 'Type'])

# 选取'Clump Thickness', 'Cell Size'两列作为特征，构建测试特征样本。
X_test = test[['Clump Thickness', 'Cell Size']]
# 选取'Type'列数据作为为测试集的正确类别答案。
y_test = test[['Type']]
```

```python
# 从sklearn.linear_model中导入LogisticRegression类，构建逻辑斯蒂回归模型用于分类任务。
from sklearn.linear_model import LogisticRegression

# 从sklearn.metrics导入accuracy_score，用来评估分类准确率。
from sklearn.metrics import accuracy_score

# 初始化一个逻辑斯蒂回归模型实例，命名为lr。
lr = LogisticRegression()

# 将X_train与y_train交给lr用于模型训练。
lr.fit(X_train, y_train)

# 将X_test交给训练好的lr模型，用于预测，预测结果存入变量y_predict。
y_predict = lr.predict(X_test)

# 对比正确类别y_test与预测类别y_predict，评估所训练模型的准确率。
accuracy_score(y_test, y_predict)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300 

test_negative = test.loc[test['Type'] == 0][['Clump Thickness', 'Cell Size', 'Type']]
test_positive = test.loc[test['Type'] == 1][['Clump Thickness', 'Cell Size', 'Type']]
```

```python
lx = np.arange(0, 12)
ly = (-lr.intercept_ - lx * lr.coef_[0,0]) / lr.coef_[0,1]
```

```python
plt.plot(lx, ly, c='black')
plt.scatter(test_negative['Clump Thickness'],test_negative['Cell Size'], marker = 'o', s=200, c='green', label='benign')
plt.scatter(test_positive['Clump Thickness'],test_positive['Cell Size'], marker = 'x', s=200, c='red', label='malignant')
plt.xlabel('Clump Thickness')
plt.ylabel('Cell Size')
plt.legend()
plt.show()
```
