```python
'''
准备数据阶段。
'''
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

train = pd.read_csv('datasets/fashionmnist/fashion-mnist_train.csv')
test = pd.read_csv('datasets/fashionmnist/fashion-mnist_test.csv')

y_train = train['label'].values
y_test = test['label'].values

X_train = train.drop('label', axis=1)
X_test = test.drop('label', axis=1)

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)

train_samples, n_features = np.shape(X_train)
test_samples = len(X_test)
```

```python
'''
构建自编码器神经网络阶段（Construction Phase)。
'''
import tensorflow as tf

BATCH_SIZE = 512
N_BATCHES = train_samples / BATCH_SIZE
N_EPOCHS = 50
LEARNING_RATE = 1e-3
INPUT_UNITS = 784
HIDDEN_UNITS = 256
ENCODED_UNITS = 64


def encoder(x):
    encoder_fc1 = tf.layers.dense(inputs=x, units=HIDDEN_UNITS, activation=tf.nn.relu, name='encoder_fc1')
    encoder_fc2 = tf.layers.dense(inputs=encoder_fc1, units=ENCODED_UNITS, activation=tf.nn.relu, name='encoder_fc2')
    return encoder_fc2


def decoder(x):
    decoder_fc1 = tf.layers.dense(inputs=x, units=HIDDEN_UNITS, activation=tf.nn.relu, name='decoder_fc1')
    decoder_fc2 = tf.layers.dense(inputs=decoder_fc1, units=INPUT_UNITS, activation=tf.nn.relu, name='decoder_fc2')
    return decoder_fc2


X = tf.placeholder(tf.float32, [None, 784])

encoded_features = encoder(X)

X_ = decoder(encoded_features)

loss = tf.reduce_mean(tf.square(X - X_), name='loss')

optimizer = tf.train.AdamOptimizer(learning_rate=LEARNING_RATE)

training_op = optimizer.minimize(loss)
```

```python
'''
执行训练网络阶段（Executation Phase)。
'''
import matplotlib.pyplot as plt
init = tf.global_variables_initializer()

with tf.Session() as sess:
    init.run()
    for iteration in range(N_EPOCHS):
        for i in range(N_BATCHES):
            np.random.seed(iteration * N_BATCHES + i)
            indices = np.random.randint(train_samples, size=BATCH_SIZE)
            X_batch = X_train[indices]
            sess.run(training_op, feed_dict={X: X_batch})
        print 'Iteration %d, loss = %f' % (iteration + 1, loss.eval(feed_dict = {X: X_train}))
        
    
    X_new = sess.run(X_, feed_dict={X: X_test[:10]})
    # Compare original images with their reconstructions
    f, a = plt.subplots(2, 10, figsize=(10, 2))
    for i in range(10):
        a[0][i].imshow(np.reshape(X_test[i], (28, 28)))
        a[1][i].imshow(np.reshape(X_new[i], (28, 28)))
    plt.show()
```
